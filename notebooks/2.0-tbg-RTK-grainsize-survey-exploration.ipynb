## Load in grain size data

### start by comparing stats for 0-line - near crest should be nearly constant

```python
def gsize_stats(fname):

    import csv

    with open(fname, 'rt') as f:

        readCSV = csv.reader(f, delimiter=',')
        skewness = []
        mean_gsize = []
        percentiles = []
        percentile_vals = []
        gsize_freqs = []
        gsize_bins = []
        sorting = []
        kurtosis = []

        for col in readCSV:

            skewness.append(col[0])
            mean_gsize.append(col[1])
            percentiles.append(col[2])
            percentile_vals.append(col[3])
            gsize_freqs.append(col[4])
            gsize_bins.append(col[5])
            sorting.append(col[6])
            kurtosis.append(col[7])
            
        return skewness, mean_gsize, percentiles, percentile_vals, gsize_freqs, gsize_bins, sorting, kurtosis

#with open('C:\\Projects\\AdvocateBeach2018\\data\\processed\\grainsize_dists\\pyfiles\\IMG_0403-cropped.csv', 'rb') as f:
#    nfile = list(csv.reader(f, delimiter=','))
```
